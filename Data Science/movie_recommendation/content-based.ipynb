```python
import pandas as pd
from sqlalchemy import create_engine
import os
from dotenv import load_dotenv
import warnings 
warnings.filterwarnings('ignore')
```

```python
load_dotenv()
user            = os.environ['usr']
password        = os.environ['pas']
host            = os.environ['hos']
port            = os.environ['por']
datascience     = os.environ['ds']

# Connect to Database

mydb = create_engine("mysql+pymysql://{usr}:{pasx}@{hosx}:{port}/{namadb}".format(
                            usr=user, pasx=password, hosx=host, port=port, namadb=datascience))

```

```python
df = pd.read_sql("select * from movie_data2 order by movie_data2.vote_count desc limit 10000", mydb)
df.head()
```

```python
# Convert atribut from string to list
df['collection'] = df['collection'].apply(lambda x:x.split(maxsplit=0))
df['genre'] = df['genre'].apply(lambda x:x.split())
df['keyword'] = df['keyword'].apply(lambda x:x.split(","))
df['company'] = df['company'].apply(lambda x:x.split(","))
df['overview'] = df['overview'].apply(lambda x:x.split())
```

```python
# Remove white spaces 
df['genre'] = df['genre'].apply(lambda x:[i.replace(" ","") for i in x])
df['keyword'] = df['keyword'].apply(lambda x:[i.replace(" ","") for i in x])
df['collection'] = df['collection'].apply(lambda x:[i.replace(" ","") for i in x])
df['company'] = df['company'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
df.head()
```

```python
movie = df.drop(columns=['release_year', 'language', 'movie_poster_path', 'runtime', 'budget', 'revenue', 'popularity', 'vote_average', 'vote_count'])
movie.head()
```

```python
movie['title'] = movie['movie_name']
movie.head()
```

```python
movie = movie.iloc[:,[0,7,1,2,3,4,5,6]]
movie.head()
```

```python
# Convert string to list
movie['movie_name'] = movie['movie_name'].apply(lambda x:x.split(maxsplit=0))

# Convert white space
movie['movie_name'] = movie['movie_name'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movie.head()
```

```python
new_df = movie.copy()
```

# Concatenating the columns into one columns

```python
new_df['tags'] = new_df['movie_name'] + new_df['genre'] + new_df['keyword'] + new_df['collection'] + new_df['company']
new_df.head()
```

```python
new_df = new_df.drop(columns=['movie_name', 'genre', 'keyword', 'collection', 'company', 'overview'])
new_df.head()
```

```python
# Convert list to string, make it to paragraph
new_df['tags'] = new_df['tags'].apply(lambda x: " ".join(x)) 
```

```python
# Convert tags into lower case alphabets
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
new_df['tags'][0]
```

# Vectorization

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer(stop_words='english')
```

```python
vector = tfidf.fit_transform(new_df['tags']).toarray()
vector
```

```python
vector.shape
```

# Convert tags with nltk

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
new_df['tags']
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
similarity.shape
```

# Find Movie Index that we search

```python
movie_index = new_df[new_df['title']=='The Maze Runner'].index[0]
movie_index
```

# Find Distances the movie to other movie using similarity for movie_index

```python
distances = similarity[movie_index]
distances
```

```python
movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[0:6]
movies_list
```

```python
for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
def recommend(movie):

    # Find the index of the movies
    movie_index = new_df[new_df['title']==movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[0:6]

    # Fetch movies from indeces
    for i in movies_list:
        print(new_df.iloc[i[0]].title)

    return 
```

```python
recommend("The Maze Runner")
```

```python
recommend("Warm Bodies")
```
